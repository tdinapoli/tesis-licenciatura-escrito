```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
periods = 4
x = np.linspace(0, periods, 1000)
y = np.sin(2*np.pi * x)

fig = plt.figure(figsize=(6,2))
plt.plot(x, y, color="k", linewidth=20)
plt.ylim([-1.5, 1.5])
plt.savefig("foton.svg", dpi=300)
plt.show()
```

```python
import scipy.stats as ss 

lambdas = np.linspace(0, 10, 1000)
excitacion = ss.norm(loc=3, scale=1).pdf(lambdas)
emision = ss.norm(loc=7, scale=1).pdf(lambdas)
excit2 = 0 #ss.norm(loc=1.5, scale=0.5).pdf(lambdas) / 4
excitacion = excitacion + excit2
emision = emision / np.max(emision)
excitacion = excitacion / np.max(excitacion)

plt.figure(figsize=(12, 5))
plt.plot(lambdas, excitacion, linewidth=3, label="Excitación",
         color="C3")
#plt.plot(lambdas, emision, linewidth=3, label="Emisión",
         #color="C0")

plt.xlim([0, 10])
plt.xticks([])
plt.yticks([])
plt.ylabel("Intensidad", fontsize=20)
plt.xlabel("$\lambda$", fontsize=20)
plt.legend(fontsize=20, loc='lower right')
plt.savefig('spec_excit.png', dpi=300)
plt.show()
```
